```python
def G_new(p, q, k, z, z_1, L_1, L_2, L_3):
   
    """

    Parameters
    ----------
    p, q: antenna indexes
    k : wavenumber
    z : 
    z_1 : z'
    L_1, L_2, L_3: list of antennas characteristic: lengths, distances, positions(ordered in the same order as the antennas in the linear array)
    Returns so called impedance kernel=Green function, used to compute the impedance coefficients
    -------
    """

    #print(L_2)
    if p == q:
        d = L_2[p]
    else:
        d = np.sqrt((L_3[p][0] - L_3[q][0])**2+ (L_3[p][1] - L_3[q][1])**2)
    #print(d)
    R = np.sqrt((z - z_1)**2 + d**2)
    if R < 1e-6:  # threshold for small distances
        return 0 
    #print(R)
    return (np.exp(-k * R*1j))/R


def Z_new(p, q, k, L_1, L_2, L_3):
    """
    Compute the coefficient of entry (p, q) of the impedance matrix (after integration by part). 
    In fact, no need to put k as argument.
    Note: The same formula can be retrieved with the near-field expression
    p, q: index of coeff Zpq in the impedance matrix

    """
    
    h_p = L_1[p]/2
    h_q = L_1[q]/2
    if p == q:
        N = 500 
    else:
        N = 300
    Z = 0
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    
    for n in range(-N//2, N//2+1):
        z_1 = (n*2*h_q)/N
        dz_1 = 2*h_q/N
        g1 = G_new(p, q, k, h_p, z_1, L_1, L_2, L_3)
        g2 = G_new(p, q, k, -h_p, z_1, L_1, L_2, L_3)
        g3 = G_new(p, q, k, 0, z_1, L_1, L_2, L_3)
        sinq = math.sin(k*(h_q- abs(z_1)))
        Z += sinq*(g1+g2-2*np.cos(k*h_p)*g3)*dz_1
    Z = Z*(complex(0, eta/(4*math.pi)))*1/(np.sin(k*h_p)*np.sin(k*h_q))          
    return Z   
```

```python
def matrixZ_new(k, L_1, L_2, L_3):
    """
    Compute the impedance matrix Z for antenna arrays. Optimized by using the symmetry of Z.
    
    """
    n = len(L_1) #nb of antennas
    #matZ = np.matrix([[Z(p,q,k) for q in range(0,n)] for p in range(0,n)])
    matZ= np.zeros((n, n), dtype=complex)
    for p in range(0, n):
        for q in range(0, p+1):
            Z_pq =  Z_new(p,q,k, L_1, L_2, L_3)
            if p == q:
                matZ[p][q] = Z_pq
            else:
                matZ[p][q] = Z_pq
                matZ[q][p] = Z_pq
    return matZ
```

```python
def input_current(V,Z):
    """
    Solve the matrix equation ZI = V, given the input voltage in each antenna and the impedance matrix.
    V: list of input voltage. Index corresponds to the antenna number in the array.
    Z: matrix with coeff obtained with matrixZ_new(k)
    """
    I = np.linalg.solve(Z, V)
    return I  
```

```python
def radiation_intensity(k, theta, phi, lengths, spacings):
    """
    Radiation intensity for the SPECIFIC case of antenna arrays along x-axis, which needs to be refined for other types.
    k: wavenumber of the emitted wave
    theta, phi: spherical coordinates
    Issue solved at theta = 0, theta = pi
    """
    def identify_vector_arg(theta, phi):
        if np.ndim(theta) > 0:
            vector_arg = theta
        else:
            vector_arg = phi
        return vector_arg

   
    wavelength = 2*math.pi/k
    n = len(lengths)
    L_1 = [l*wavelength for l in lengths]
    L_2 = [0.001*wavelength for _ in range(n)]
    L_3 = [(x*wavelength, y*wavelength, z*wavelength) for (x, y, z) in spacings] #L_3 = [(sum(spacings[:i])*wavelength, 0) for i in range(n)]
    
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    V = [0, 0, 0, 0, 0, 1]+[0]*(n-6) #[0, 1]+[0]*(n-2)
    I = input_current(V,matrixZ_new(k, L_1, L_2, L_3))
    M = identify_vector_arg(theta, phi)
    
    U = np.zeros_like(M, dtype=complex)  # to accumulate field
    for p in range(0, len(L_1)):
        h_p = L_1[p]/2
        term1 = np.exp(1j*k*np.sin(theta)*(L_3[p][0]*np.cos(phi)+L_3[p][1]*np.sin(phi)))
        if np.ndim(theta) > 0:
            mask = (theta != 0) & (theta != np.pi) #to deal with division by 0 at theta = 0 and theta = pi: extension to 0 by continuity
            term2 = np.zeros_like(theta, dtype=complex)
            term2[mask] = (np.cos(k*h_p*np.cos(theta[mask]))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta[mask]))
        else:
            term2 =  (np.cos(k*h_p*np.cos(theta))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta))
        U += I[p]*term1*term2
    return (eta/(8*np.pi**2))*abs(U)**2
```

```python
def directors(positions, lengths, n, d, l):
    """
    n: number of director desired
    positions, lengths: list of reflectors and driven element positions and lengths to be updated
    d : distance between directors
    l : difference of length between directors
    """
    x = positions[-1][0]
    pos = positions + [(x+i*d, 0, 0) for i in range(n)]
    leng = lengths + [lengths[-1]-0.04-i*l for i in range(n)]
    return pos, leng
```

```python
def radiated_power(k, lengths, spacings):
    theta = np.linspace(0, np.pi, 200)
    phi = np.linspace(0, 2 * np.pi, 200)
    dtheta = theta[1] - theta[0]
    dphi = phi[1] - phi[0]
    
    THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
    U = radiation_intensity(k, THETA, PHI, lengths, spacings)
    
    integrand = U * np.sin(THETA)
    P_rad = np.sum(integrand) * dtheta * dphi
    return P_rad
```

```python
def directivity(k, theta, phi, lengths, spacings):


    # Compute radiation intensity U(, )
    U = radiation_intensity(k, theta, phi, lengths, spacings)  # shape (len(theta), len(phi))
    #print("U shape:", U.shape)
    P = radiated_power(k, lengths, spacings)
    
    
    # Directivity
    D = 4 * np.pi * U / P
    if np.any(D < 0):
        raise ValueError("Directivity contains negative values. Check radiation intensity calculation.")
    #A negative dB simply means the directivity is less than 1 
    #(i.e., it's radiating less intensely in the peak direction than an isotropic radiator would).
    #In the case of Yagi, it should not happen along positive x.
    return D
```

```python
def max_directivity(k, lengths, spacings):
    phi = np.linspace(0, 2 * np.pi, 500)
    theta = np.linspace(0, np.pi, 500)

    # Create meshgrid
    THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
    D = directivity(k, THETA, PHI, lengths, spacings)
    D_max = np.max(D)
    max_index = np.unravel_index(np.argmax(D), D.shape)
    
    max_theta = THETA[max_index]
    max_phi = PHI[max_index]
    return f'The maximum directivity is {D_max} for theta, phi = {max_theta}, {max_phi}'
```

```python
import random
import matplotlib.pyplot as plt
import time

# Algorithme gntique personnalis pour optimiser l'antenne Yagi
def optimize_yagi_antenna(n_elements=5, population_size=100, generations=50):
    """
    Optimise une antenne Yagi sans utiliser la bibliothque DEAP.
    
    Args:
        n_elements: Nombre d'lments dans l'antenne Yagi
        population_size: Taille de la population
        generations: Nombre de gnrations
        
    Returns:
        Tuple contenant (meilleures longueurs, meilleurs espacements, directivit maximale)
    """
    global l_1, l_2
    
    # Configuration pour l'optimisation
    # Bornes typiques pour les antennes Yagi (en longueurs d'onde)
    LENGTH_MIN, LENGTH_MAX = 0.40, 0.65  # Longueurs typiques
    REFLECTOR_MIN, REFLECTOR_MAX = 0.50, 0.65  # Rflecteur gnralement plus long
    DIPOLE_MIN, DIPOLE_MAX = 0.45, 0.52  # Diple autour de 0.5
    DIRECTOR_MIN, DIRECTOR_MAX = 0.40, 0.48  # Directeurs gnralement plus courts
    SPACING_MIN, SPACING_MAX = 0.15, 0.40  # Espacements typiques
    
    # Nombre total de paramtres: n_elements longueurs + (n_elements-1) espacements
    n_params = 2 * n_elements - 1
    
    # Cration d'un individu alatoire avec des contraintes spcifiques
    def create_individual():
        individual = []
        
        # Longueurs des lments
        individual.append(random.uniform(REFLECTOR_MIN, REFLECTOR_MAX))  # Rflecteur
        individual.append(random.uniform(DIPOLE_MIN, DIPOLE_MAX))  # Diple
        
        # Directeurs (peuvent tre de plus en plus courts)
        for i in range(2, n_elements):
            individual.append(random.uniform(DIRECTOR_MIN, DIRECTOR_MAX))
        
        # Espacements entre les lments
        for i in range(n_elements - 1):
            individual.append(random.uniform(SPACING_MIN, SPACING_MAX))
            
        return individual
    
    # Cration de la population initiale
    population = [create_individual() for _ in range(population_size)]
    
    # Fonction d'valuation
    def evaluate(individual):
        lengths = individual[:n_elements]
        spacings = individual[n_elements:]

        k = 2 * np.pi

        # Utiliser une rsolution plus basse pendant l'optimisation
        # Rduire significativement le nombre de points
        #phi = np.linspace(0, 2 * np.pi, 36)  # Rduire de 72  36 points
        #theta = np.linspace(0, np.pi, 36)    # Rduire de 72  36 points

        #THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
        phi = np.array([0])  # Direction de l'axe principal
        theta = np.linspace(15*np.pi/32, 17*np.pi/32, 36)  # Autour du plan H

        try:
            D = directivity(k, theta, phi, lengths, spacings)
            D_max = np.max(D)

            if not np.isfinite(D_max) or D_max <= 0:
                return 0.0

            return D_max
        except Exception as e:
            return 0.0
    
    # Slection par tournoi
    def tournament_selection(population, fitnesses, tournament_size=3):
        selected = []
        for _ in range(len(population)):
            # Slection alatoire des participants au tournoi
            aspirants_idx = random.sample(range(len(population)), tournament_size)
            # Le gagnant est celui avec la meilleure fitness
            winner_idx = max(aspirants_idx, key=lambda i: fitnesses[i])
            selected.append(population[winner_idx].copy())
        return selected
    
    # Croisement mixte (blend crossover)
    def blend_crossover(parent1, parent2, alpha=0.5):
        child1 = []
        child2 = []
        
        for gene1, gene2 in zip(parent1, parent2):
            # Dterminer les bornes du mlange
            lower = min(gene1, gene2) - alpha * abs(gene2 - gene1)
            upper = max(gene1, gene2) + alpha * abs(gene2 - gene1)
            
            # Gnrer les nouvelles valeurs
            value1 = random.uniform(lower, upper)
            value2 = random.uniform(lower, upper)
            
            child1.append(value1)
            child2.append(value2)
            
        return child1, child2
    
    # Mutation gaussienne
    def gaussian_mutation(individual, mu=0, sigma=0.05, prob=0.2):
        for i in range(len(individual)):
            if random.random() < prob:
                individual[i] += random.gauss(mu, sigma)
                
                # Appliquer des limites spcifiques selon le type de paramtre
                if i < n_elements:  # Si c'est une longueur
                    if i == 0:  # Rflecteur
                        individual[i] = max(REFLECTOR_MIN, min(REFLECTOR_MAX, individual[i]))
                    elif i == 1:  # Diple
                        individual[i] = max(DIPOLE_MIN, min(DIPOLE_MAX, individual[i]))
                    else:  # Directeurs
                        individual[i] = max(DIRECTOR_MIN, min(DIRECTOR_MAX, individual[i]))
                else:  # Si c'est un espacement
                    individual[i] = max(SPACING_MIN, min(SPACING_MAX, individual[i]))
        
        return individual
    
    # Variables pour le suivi de la progression
    best_individual = None
    best_fitness = -float('inf')
    avg_fitnesses = []
    max_fitnesses = []
    
    # Dmarrage du chronomtre
    start_time = time.time()
    
    # Boucle principale de l'algorithme gntique
    print("Dmarrage de l'optimisation...")
    for gen in range(generations):
        # valuation de la population
        fitnesses = [evaluate(ind) for ind in population]
        
        # Mise  jour du meilleur individu
        current_best_idx = np.argmax(fitnesses)
        current_best_fitness = fitnesses[current_best_idx]
        
        if current_best_fitness > best_fitness:
            best_fitness = current_best_fitness
            best_individual = population[current_best_idx].copy()
            
        # Statistiques
        avg_fitness = np.mean(fitnesses)
        max_fitness = np.max(fitnesses)
        avg_fitnesses.append(avg_fitness)
        max_fitnesses.append(max_fitness)
        
        # Affichage de la progression
        if gen % 10 == 0 or gen == generations - 1:
            elapsed = time.time() - start_time
            print(f"Gnration {gen}/{generations} - Meilleure: {max_fitness:.2f}, "
                  f"Moyenne: {avg_fitness:.2f}, Temps coul: {elapsed:.1f}s")
        
        # Slection
        selected = tournament_selection(population, fitnesses)
        
        # Nouvelle population
        new_population = []
        
        # Elitisme: garder le meilleur individu
        new_population.append(best_individual.copy())
        
        # Remplir le reste de la population avec croisement et mutation
        while len(new_population) < population_size:
            # Slection des parents
            parent1 = random.choice(selected)
            parent2 = random.choice(selected)
            
            # Croisement avec une probabilit de 0.7
            if random.random() < 0.7:
                child1, child2 = blend_crossover(parent1, parent2)
            else:
                child1, child2 = parent1.copy(), parent2.copy()
            
            # Mutation avec une probabilit de 0.2
            child1 = gaussian_mutation(child1)
            child2 = gaussian_mutation(child2)
            
            # Ajout  la nouvelle population
            new_population.append(child1)
            if len(new_population) < population_size:
                new_population.append(child2)
        
        # Mise  jour de la population
        population = new_population
    
    # Fin de l'optimisation
    total_time = time.time() - start_time
    print(f"\nOptimisation termine en {total_time:.1f} secondes")
    
    # Sparation des longueurs et espacements du meilleur individu
    best_lengths = best_individual[:n_elements]
    best_spacings = best_individual[n_elements:]
    
    # Affichage des rsultats
    print("\nMeilleure solution trouve:")
    print(f"Directivit maximale: {best_fitness:.2f}")
    print(f"Longueurs optimales: {[f'{l:.3f}' for l in best_lengths]}")
    print(f"Espacements optimaux: {[f'{s:.3f}' for s in best_spacings]}")
    
    # Tracer l'volution de la fitness
    plt.figure(figsize=(10, 6))
    plt.plot(range(generations), max_fitnesses, 'b-', label='Maximum')
    plt.plot(range(generations), avg_fitnesses, 'r-', label='Moyenne')
    plt.title('volution de la directivit au cours des gnrations')
    plt.xlabel('Gnration')
    plt.ylabel('Directivit')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Reprsentation visuelle de l'antenne Yagi optimise
    plt.figure(figsize=(12, 5))
    x_positions = [0]  # Position du rflecteur
    element_names = ["Rflecteur", "Diple"]
    
    # Ajouter les positions des autres lments
    for i in range(len(best_spacings)):
        x_positions.append(x_positions[-1] + best_spacings[i])
        if i >= 1:
            element_names.append(f"Directeur {i}")
    
    # Tracer chaque lment
    for i, (pos, length, name) in enumerate(zip(x_positions, best_lengths, element_names)):
        plt.plot([pos, pos], [-length/2, length/2], 'b-', linewidth=3)
        plt.text(pos, -length/2 - 0.05, name, ha='center')
    
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axis('equal')
    plt.grid(True)
    plt.title('Schma de l\'antenne Yagi optimise')
    plt.xlabel('Position ()')
    plt.ylabel('Longueur ()')
    plt.xlim(-0.1, x_positions[-1] + 0.1)
    plt.show()
    
    # Vrifier la directivit maximale avec une rsolution plus fine
    k = 2 * np.pi
    result = max_directivity(k, best_lengths, best_spacings)
    print("\nVrification de la directivit maximale avec une rsolution plus fine:")
    print(result)
    
   


```

```python
def G_2(p, q, k, z, z_1, L_1, L_2, L_3):
    """
    Green's function kernel for impedance calculation.
    z: observation point (scalar), z_1: integration points (array)
    """
    if p == q:
        d = L_2[p] 
    else:
        d = np.linalg.norm(np.array(L_3[p]) - np.array(L_3[q]))

    R = np.sqrt((z - z_1)**2 + d**2)
    
        # Avoid division by zero with np.where
    #G = np.where(R < 1e-10, 0, np.exp(-1j * k * R) / R)
    G = np.exp(-1j * k * R) / R
    return G
```

```python
def G_3(p, q, k, z, z_1, L_1, L_2, L_3):
    if p == q:
        theta = np.linspace(0, np.pi/2, 100)                  # (T,)
        theta = theta[None, None, :]                          # reshape to (1, 1, T)
        a_p =  L_2[p] 
        z = np.atleast_1d(z)[:, None, None]                   # shape (N, 1, 1)
        z_1 = np.atleast_1d(z_1)[None, :, None]       # shape (1, M, 1)

        delta_z = z - z_1                                 # shape (N, M, 1)

        sin_theta_sq = np.sin(theta)**2                       # shape (1, 1, T)

        R_sq = delta_z**2 + 4*a_p**2 - 4*a_p**2 * sin_theta_sq
        R_sq = np.maximum(R_sq, 1e-12)                        # Avoid division by 0
        R = np.sqrt(R_sq)

        G_integrand = np.exp(-1j * k * R) / R                 # shape (N, M, T)

        integral = np.trapz(G_integrand, theta, axis=2)       # shape (N, M)

        G = (2 / np.pi) * integral      
    else:
        d = np.linalg.norm(np.array(L_3[p]) - np.array(L_3[q]))
        R = np.sqrt((z - z_1)**2 + d**2)
        G = np.exp(-1j * k * R) / R

    return G
```

```python
M = 40#we divide each antenna into N segment
N = 2*M+1
#V = [0]*N+[1]*N+[0]*N #input voltage



    
def Z_mn(p, q, z_m, z_n, delta_q, L_1, L_2, L_3, k):
    """
    Computes the integrated MoM matrix element Z_mn using pulse basis and point matching.
    Integration is over segment n, centered at z_n, evaluated at z_m.
    """
    eta = 376.7
    # Integration bounds for segment n
    z_start = - delta_q / 2 #z_n - delta_n / 2
    z_end = delta_q / 2 #z_n + delta_n / 2

    # Use numerical integration (trapezoidal rule or similar)
    z_prime = np.linspace(z_start, z_end, 15)
    G =  G_3(p, q, k, z_m-z_n, z_prime, L_1, L_2, L_3)
    
    # Approximate integral
    integral = np.trapz(G, z_prime)
    return (1j*eta)/(2*np.pi)*integral

def submatrix(p, q, z_p, delta_p, z_q, delta_q, L_1, L_2, L_3, k):
    """
    z_p, z_q: arrays of segment centers for antennas p and q
    p, q: antennas p and q
    L_1, L_2, L_3: list of antennas characteristic: lengths, distances, positions
    Returns: NxN matrix Z_pq
    """
    N = len(z_p)
    Z_pq = np.zeros((N, N), dtype=complex)
    for m in range(N):
        for n in range(N):
            Z_pq[m, n] = Z_mn(p, q, z_p[m], z_q[n], delta_q, L_1, L_2, L_3, k)
    return Z_pq

def matrix_Z(L_1, L_2, L_3, k):
    """
    all_z: list of arrays [z_0, z_1, ..., z_K] of segment centers for each antenna
    all_d: 2D array of mutual distances d_pq between antennas (size (K+1)x(K+1))
    Returns: full MoM matrix Z
    """
    global N
    
    K_plus_1 = len(L_1)
    Z = np.zeros((K_plus_1 * N, K_plus_1 * N), dtype=complex)

    for p in range(K_plus_1):
        h_p = L_1[p]/2
        delta_p = 2*h_p/N #segment width
        z_p = [-h_p+(n-0.5)*delta_p for n in range(N)] #list of segment centers
        for q in range(K_plus_1):
            h_q = L_1[q]/2
            delta_q = 2*h_q/N #segment width
            z_q = [-h_q+(n-0.5)*delta_q for n in range(N)] #list of segment centers
            Z_pq = submatrix(p, q, z_p, delta_p, z_q, delta_q, L_1, L_2, L_3, k)
            Z[p*N:(p+1)*N, q*N:(q+1)*N] = Z_pq
    return Z

```

```python
def radiation_intensity(k, theta, phi, lengths, positions):
    """
    Radiation intensity for the SPECIFIC case of antenna arrays along x-axis, which needs to be refined for other types.
    k: wavenumber of the emitted wave
    theta, phi: spherical coordinates
    Issue solved at theta = 0, theta = pi
    """
    def identify_vector_arg(theta, phi):
        if np.ndim(theta) > 0:
            vector_arg = theta
        else:
            vector_arg = phi
        return vector_arg

   
    wavelength = 2*math.pi/k
    n = len(lengths)
    L_1 = [l*wavelength for l in lengths]
    L_2 = [0.001*wavelength for _ in range(n)]
    L_3 = [(x*wavelength, y*wavelength, z*wavelength) for (x, y, z) in positions] #L_3 = [(sum(spacings[:i])*wavelength, 0) for i in range(n)]
    
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    V = [0, 1]+[0]*(n-2) #[0, 1]+[0]*(n-2)
    I = input_current(V,matrixZ_new(k, L_1, L_2, L_3))
    M = identify_vector_arg(theta, phi)
    
    U = np.zeros_like(M, dtype=complex)  # to accumulate field
    for p in range(0, len(L_1)):
        h_p = L_1[p]/2
        term1 = np.exp(1j*k*np.sin(theta)*(L_3[p][0]*np.cos(phi)+L_3[p][1]*np.sin(phi)))
        if np.ndim(theta) > 0:
            mask = (theta != 0) & (theta != np.pi) #to deal with division by 0 at theta = 0 and theta = pi: extension to 0 by continuity
            term2 = np.zeros_like(theta, dtype=complex)
            term2[mask] = (np.cos(k*h_p*np.cos(theta[mask]))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta[mask]))
        else:
            term2 =  (np.cos(k*h_p*np.cos(theta))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta))
        U += I[p]*term1*term2
    return (eta/(8*np.pi**2))*abs(U)**2




def radiation_MoM(k, theta, phi, lengths, positions):
    global N
    def identify_vector_arg(theta, phi):
        if np.ndim(theta) > 0:
            vector_arg = theta
        else:
            vector_arg = phi
        return vector_arg

   
    wavelength = 2*math.pi/k
    n = len(lengths)
    L_1 = [l*wavelength for l in lengths]
    L_2 = [0.003*wavelength for _ in range(n)]
    L_3 = [(x*wavelength, y*wavelength, z*wavelength) for (x, y, z) in positions] #L_3 = [(sum(spacings[:i])*wavelength, 0) for i in range(n)]
    
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    V = [0]*N+[0]*(N//2)+[1]+[0]*(N//2)+[0]*N #input voltage
    I = input_current(V,matrix_Z(L_1, L_2, L_3, k))
    print(matrix_Z(L_1, L_2, L_3, k)[4, 1])
    M = identify_vector_arg(theta, phi)
    F = np.zeros_like(M, dtype=complex)  # to accumulate field
    for p in range(0, len(L_1)):
        #reconstruct I_p
        h_p = L_1[p]/2
        delta_p = 2*h_p/N #segment width
        z_p = [-h_p+(n-0.5)*delta_p for n in range(N)] #list of segment centers
        k_z = k*np.cos(theta)
        current = I[p*N: p*N+N]
        h_p = L_1[p]/2
        term1 = np.exp(1j*k*np.sin(theta)*(L_3[p][0]*np.cos(phi)+L_3[p][1]*np.sin(phi)))
        term2 = calculate_sum_integral(z_p, delta_p, N, current, k_z)
            
        F += term1*term2
    return (eta*k**2/(32*np.pi**2))*abs(np.sin(theta)*F)**2
```

```python
def MoM_plot(lengths, positions):
    theta = np.pi/2
    phi = np.linspace(0, 2*np.pi, 200)
    k = 1

    U = radiation_intensity(k, theta, phi, lengths, positions)
    U_MoM = radiation_MoM(k, theta, phi, lengths, positions)
    # Normalize
    U /= np.max(U)
    U_MoM /= np.max(U_MoM)
    
    # Polar plot
    plot_vertical_lines(positions, lengths, l=1.0, color='b')
    plt.figure(figsize=(6, 6))
    ax = plt.subplot(111, projection='polar')
    ax.plot(phi, U, 'b-')
    ax.plot(phi, U_MoM, 'r--')
    ax.set_title("MoM vs EMF (azimuthal angle)", va='bottom')
    ax.set_rticks([0.25, 0.5, 0.75, 1.0])
    plt.show()
```

```python
def calculate_sum_integral(z_p, delta, N, I, k_z):
    """
    Calculate the sum of integrals:
    sum_{i=1}^N int_{z_p[i]-delta/2}^{z_p[i]+delta/2} I[i] e^{jk_z z'} dz'
    
    Parameters:
    z_p (list/array): List of position values
    delta (float): Width of integration region
    N (int): Number of segments
    I (list/array): Array of coefficients
    k_z (float/complex): Wave number in z-direction
    
    Returns:
    complex: Result of the summation of integrals
    """
    # Initialize result
    result = 0j
    
    # For each segment
    for i in range(N):
        # Integration limits
        z_start = z_p[i] - delta/2
        z_end = z_p[i] + delta/2
        
        # Analytical solution of the integral:
        # I[i] * int_{z_start}^{z_end} e^{jk_z z'} dz'
        # = I[i] * [e^{jk_z z'} / (jk_z)]_{z_start}^{z_end}
        # = I[i] * (e^{jk_z*z_end} - e^{jk_z*z_start}) / (j*k_z)
        
        if abs(k_z) < 1e-10:  # Handle case when k_z is close to zero
            integral = I[i] * (z_end - z_start)
        else:
            integral = I[i] * (np.exp(1j * k_z * z_end) - np.exp(1j * k_z * z_start)) / (1j * k_z)
            
        result += integral
    
    return result
```
