## Create your Yagi antenna

```python
class Antenna:
    def __init__(self, length, radius, position):
        """
        Initialize an Antenna object.

        Parameters:
        - length (float): The length of the antenna.
        - radius (float): The radius of the antenna.
        - position (tuple of 2 float): The (x, y) position of the antenna in space. 
        (assuming the antenna lies on the plane z=0)
        """
        self.length = length
        self.radius = radius
        self.position = position  # Expected to be a 2D coordinate: (x, y) 

    def __repr__(self):
        return (f"Antenna(length={self.length}, radius={self.radius}, "
                f"position={self.position})")

class Array:
    
    def __init__(self):
        """
        Initialize an empty antenna array.
        """
        self.antennas = []
        self.lengths = []
        self.radii = []
        self.positions = []

    def add_antenna(self, antenna):
        """
        Add an Antenna object to the array.

        Parameters:
        - antenna (Antenna): The antenna to add.
        """
        if not isinstance(antenna, Antenna):
            raise TypeError("Only Antenna objects can be added to the array.")
        self.antennas.append(antenna)
        self.lengths.append(antenna.length)
        self.radii.append(antenna.radius)
        self.positions.append(antenna.position)
    
    def remove_antenna(self):
        self.antennas.pop()
        self.lengths.pop()
        self.radii.pop()
        self.positions.pop()

    def __len__(self):
        return len(self.antennas)

    def __getitem__(self, index):
        return self.antennas[index]

    def __repr__(self):
        return f"Array with {len(self.antennas)} antennas:\n" + "\n".join(str(a) for a in self.antennas)
    
   
def create_array(lengths, r, sep, n):
    """
    Create a Yagi antenna with:
    lengths: array of 3 element with the length of each element in the antenna: reflector, driven/feeder, directors
    r: antenna radii (same for each element)
    sep: separation between antennas
    !Note: All lengths given in units of the wavelength lambda!
    n: number of antenna in the array
    
    return three lists describing the Yagi array
    """
    array = Array()
    l_1 = [lengths[0], length[1]] + [lengths[2]]*(n-2)
    l_2 = [r]*n
    l_3 = [(i*sep,0) for i in range(n)]
    for k in range(n):                #useless for now but could be used later on
        antenna = Antenna(l_1[k], l_2[k], l_3[k])
        array.add(antenna)
        
    return l_1, l_2, l_3
    
```

# Computing the Intensity Profile of a Yagi Uda Antenna

```python
"All lengths given in units of the wavelength lambda"
l_1 = [0.50, 0.48, 0.46] #[0.6, 0.5, 0.45, 0.45, 0.45] #antenna lengths. 
l_2 = [0.003,0.003,0.003] #[0.001, 0.001, 0.001, 0.001, 0.001] #antenna radii
l_3 = [(-0.125,0), (0,0), (0.125,0)] #[(0,0), (0.5,0), (1,0), (1.5,0), (2,0)] #antenna position

V = [0, 1]+[0]*(len(l_1)-2) #voltage in each antenna =[0, 1, 0, 0, 0] for 5 elements 

def G_new(p, q, k, z, z_1, L_1, L_2, L_3):
   
    """

    Parameters
    ----------
    p, q: antenna indexes
    k : wavenumber
    z : 
    z_1 : z'
    L_1, L_2, L_3: list of antennas characteristic (ordered in the same order as the antennas in the linear array)
    Returns so called impedance kernel=Green function, used to compute the impedance coefficients
    -------
    """

    #print(L_2)
    if p == q:
        d = L_2[p]
    else:
        d = math.sqrt((L_3[p][0] - L_3[q][0])**2+ (L_3[p][1] - L_3[q][1])**2)
    #print(d)
    R = math.sqrt((z - z_1)**2 + d**2)
    if R < 1e-6:  # threshold for small distances
        return 0 
    #print(R)
    return (cmath.exp(-k * R*1j))/R


def Z_new(p, q, k=1):
    """
    Compute the coefficient of entry (p, q) of the impedance matrix (after integration by part). 
    In fact, no need to put k as argument.
    Note: The same formula can be retrieved with the near-field expression
    p, q: index of coeff Zpq in the impedance matrix

    """
    global l_1, l_2, l_3
    wavelength = 2*math.pi/k
    L_1 = [l*wavelength for l in l_1]
    L_2 = [l*wavelength for l in l_2]
    L_3 = [(x*wavelength, y*wavelength) for (x, y) in l_3]
    h_p = L_1[p]/2
    h_q = L_1[q]/2
    if p == q:
        N = 500 
    else:
        N = 300
    Z = 0
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    
    for n in range(-N//2, N//2+1):
        z_1 = (n*2*h_q)/N
        dz_1 = 2*h_q/N
        g1 = G_new(p, q, k, h_p, z_1, L_1, L_2, L_3)
        g2 = G_new(p, q, k, -h_p, z_1, L_1, L_2, L_3)
        g3 = G_new(p, q, k, 0, z_1, L_1, L_2, L_3)
        sinq = math.sin(k*(h_q- abs(z_1)))
        Z += sinq*(g1+g2-2*np.cos(k*h_p)*g3)*dz_1
    Z = Z*(complex(0, eta/(4*math.pi)))*1/(np.sin(k*h_p)*np.sin(k*h_q))          
    return Z   
```

```python
def matrixZ_new(k):
    """
    Compute the impedance matrix Z for antenna arrays. Optimized by using the symmetry of Z.
    
    """
    n = len(l_1) #nb of antennas
    #matZ = np.matrix([[Z(p,q,k) for q in range(0,n)] for p in range(0,n)])
    matZ= np.zeros((n, n), dtype=complex)
    for p in range(0, n):
        for q in range(0, p+1):
            Z_pq =  Z_new(p,q,k)
            if p == q:
                matZ[p][q] = Z_pq
            else:
                matZ[p][q] = Z_pq
                matZ[q][p] = Z_pq
    return matZ
```

```python
def input_current(V,Z):
    """
    Solve the matrix equation ZI = V, given the input voltage in each antenna and the impedance matrix.
    V: list of input voltage. Index corresponds to the antenna number in the array.
    Z: matrix with coeff obtained with matrixZ_new(k)
    """
    I = np.linalg.solve(Z, V)
    return I  
```

```python
def radiation_intensity(k, theta, phi):
    """
    Radiation intensity for the SPECIFIC case of antenna arrays along x-axis, which needs to be refined for other types.
    k: wavenumber of the emitted wave
    theta, phi: spherical coordinates
    Issue solved at theta = 0, theta = pi
    """
    def identify_vector_arg(theta, phi):
        if np.ndim(theta) > 0:
            vector_arg = theta
        else:
            vector_arg = phi
        return vector_arg

    global l_1, l_2, l_3
    wavelength = 2*math.pi/k
    L_1 = [l*wavelength for l in l_1]
    L_2 = [l*wavelength for l in l_2]
    L_3 = [(x*wavelength, y*wavelength) for (x, y) in l_3]
    
    eta = 376.7 #sqrt(mu_0/epsilon_0)
    I = input_current(V,matrixZ_new(k))
    M = identify_vector_arg(theta, phi)
    
    U = np.zeros_like(M, dtype=complex)  # to accumulate field
    for p in range(0, len(L_1)):
        h_p = L_1[p]/2
        term1 = np.exp(1j*k*np.sin(theta)*L_3[p][0]*np.cos(phi))
        if np.ndim(theta) > 0:
            mask = (theta != 0) & (theta != np.pi) #to deal with division by 0 at theta = 0 and theta = pi: extension to 0 by continuity
            term2 = np.zeros_like(theta, dtype=complex)
            term2[mask] = (np.cos(k*h_p*np.cos(theta[mask]))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta[mask]))
        else:
            term2 =  (np.cos(k*h_p*np.cos(theta))-np.cos(k*h_p))/(np.sin(k*h_p)*np.sin(theta))
        U += I[p]*term1*term2
    return (eta/(8*np.pi**2))*abs(U)**2

def radiation_intensity_db(k, theta, phi):
    """
    Radiation intensity in dB 
    """
    return 10*np.log10(radiation_intensity(k, theta, phi))

def radiation_theta(k, phi):
    """
    Plot radiation intensity versus theta (spherical coordinates)
    """
    theta = np.linspace(0, 2*np.pi, 500)
    U = radiation_intensity(k, theta, phi)
    
     # Plot 
    plt.plot(theta, U)
    plt.xlabel("theta")
    plt.ylabel("U")
    plt.title("Radiation intensity U(theta)")
    plt.grid(True)
    plt.show()
    
    # Polar plot
    plt.figure()
    ax = plt.subplot(111, projection='polar')
    ax.plot(theta, U)

    ax.set_title("Radiation intensity U(θ)", va='bottom')
    ax.set_theta_zero_location("N")   # Zero at the top (θ = 0)
    ax.set_theta_direction(1)         # Counter-clockwise for θ

def radiation_phi(k, theta):
    """
    Plot radiation intensity versus the azimuthal angle phi (spherical coordinates)
    """
    phi = np.linspace(-np.pi, np.pi, 500)
    U = radiation_intensity(k, theta, phi)
    
    # Plot 
    plt.plot(phi, U)
    plt.xlabel("phi")
    plt.ylabel("U")
    plt.title("Radiation intensity U(phi)")
    plt.grid(True)
    plt.show()
    
    # Polar plot
    plt.figure()
    ax = plt.subplot(111, projection='polar')
    ax.plot(phi, U)

    ax.set_title("Radiation intensity U(φ)", va='bottom')
    ax.set_theta_zero_location("N")  # Zero at the top
    ax.set_theta_direction(-1)       # Clockwise

    plt.show()

```

```python
def radiated_power(k):
    theta = np.linspace(0, np.pi, 200)
    phi = np.linspace(0, 2 * np.pi, 200)
    dtheta = theta[1] - theta[0]
    dphi = phi[1] - phi[0]
    
    THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
    U = radiation_intensity(k, THETA, PHI)
    
    integrand = U * np.sin(THETA)
    P_rad = np.sum(integrand) * dtheta * dphi
    return P_rad
```

 # Directivity

```python
def directivity(k, theta, phi):


    # Compute radiation intensity U(θ, φ)
    U = radiation_intensity(k, theta, phi)  # shape (len(theta), len(phi))
    #print("U shape:", U.shape)
    P = radiated_power(k)
    
    
    # Directivity
    D = 4 * np.pi * U / P
    if np.any(D < 0):
        raise ValueError("Directivity contains negative values. Check radiation intensity calculation.")
    #A negative dB simply means the directivity is less than 1 
    #(i.e., it's radiating less intensely in the peak direction than an isotropic radiator would).
    #In the case of Yagi, it should not happen along positive x.
    return D

def max_directivity(k):
    phi = np.linspace(0, 2 * np.pi, 500)
    theta = np.linspace(0, np.pi, 500)

    # Create meshgrid
    THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
    D = directivity(k, THETA, PHI)
    D_max = np.max(D)
    max_index = np.unravel_index(np.argmax(D), D.shape)
    
    max_theta = THETA[max_index]
    max_phi = PHI[max_index]
    return f'The maximum directivity is {D_max} for theta, phi = {max_theta}, {max_phi}'

def directivity_dB(k, theta, phi):
    D = directivity(k, theta, phi)
    return 10*np.log10(D)

def plot_directivity(k):
    """
    Plot 3D directivity pattern with colorbar (red = higher directivity)
    """
    phi = np.linspace(0, 2 * np.pi, 500)
    theta = np.linspace(0, np.pi, 500)


    # Create meshgrid
    THETA, PHI = np.meshgrid(theta, phi, indexing='ij')
    
    # Calculate directivity D(θ, φ)
    D = directivity(k, THETA, PHI)  # shape (len(theta), len(phi))

    # Convert to Cartesian coordinates
    R = D / np.max(D)  # Normalize for nicer plotting
    X = R * np.sin(THETA) * np.cos(PHI)
    Y = R * np.sin(THETA) * np.sin(PHI)
    Z = R * np.cos(THETA)

    # Create the 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface with color representing directivity
    surf = ax.plot_surface(X, Y, Z, facecolors=cm.plasma(R), rstride=1, cstride=1, antialiased=False, linewidth=0)

    # Add colorbar
    mappable = cm.ScalarMappable(cmap=cm.plasma)
    mappable.set_array(R)
    cbar = plt.colorbar(mappable, ax=ax, shrink=0.6, pad=0.1)
    cbar.set_label('Normalized Directivity')

    
    # Add axis labels
    ax.set_xlabel('X Axis')
    ax.set_ylabel('Y Axis')
    ax.set_zlabel('Z Axis')

    ax.set_title("3D Radiation Pattern (Directivity)")
    plt.show()
```
